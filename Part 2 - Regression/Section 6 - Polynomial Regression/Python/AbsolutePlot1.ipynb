```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.DataFrame(np.arange(-20, 21, 1), columns = ['X'])
```

```python
df.head(10)
```

```python
df['y'] = df.X.apply(lambda X: abs(np.sqrt((3-2*X)**2)))
```

```python
df
```

```python
df.plot.scatter(x = 'X', y = 'y', grid=True)
sns.lineplot(df, x='X', y='y', marker='+')
plt.axhline(y=5, c='g', linestyle='--')
plt.axvline(x = 1.5, color='r', linestyle='--')
plt.ylabel('y=|sqrt(3-2X)^2|')
plt.savefig('np.sqrt_3-2X_.png')
```
